```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

x=np.linspace(0,20,num=200,endpoint=True)
y0=np.sin(x)*4/math.pi
y1=np.sin(3*x)*4/(3*math.pi)
```

```python
y=[ ]
for i in range(1000):
    temp = np.sin((2*i+1)*x)*4/((2*i+1)*math.pi)
    y.append ( temp )
```

```python
plt.figure(figsize=(15,10), dpi=50)
for i in range(5):
    plt.subplot(5, 1, i+1)
    plt.plot(x, y[i])
```

```python
def sum_vector(y,n):
    temp = np.zeros_like(y[0])
    for i in range(n+1):
        temp += y[i]
    return temp


n=10
plt.figure(figsize=(15,n*2), dpi=50)
for i in range(10):
    plt.subplot(10, 1, i+1)
    plt.plot(x, sum_vector(y,i))
```

##  
$\displaystyle f(x)=\frac{\pi}{2} + \sum_{(n=1)}^{k} \left( - \frac{4}{(2n-1)^2\pi}\right)cos((2n-1)x)$


```python
x=np.linspace(-10,10,num=200,endpoint=True)

def triangle_wave(x,k):
    temp = np.zeros_like(x)
    for i in range(1,k+1):
        temp += np.cos((2*i-1)*x)* (-4)/(math.pi * (2*i-1)**2)
    return temp#+math.pi/2

n=4
plt.figure(figsize=(10,n*3), dpi=80)
for i in range(1,n+1):
    plt.subplot(n, 1, i)
    plt.plot(x, triangle_wave(x,i))
    

plt.figure(figsize=(10,3), dpi=80)
plt.plot(x, triangle_wave(x,20))    
```
